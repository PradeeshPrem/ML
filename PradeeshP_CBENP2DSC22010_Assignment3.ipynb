```python
df.columns = ['Rank','Country','Life expectancy at birth(years)','Expected years of schooling(years)','Mean years of schooling(years)','Gross national income (GNI) per capita']
```

```python
df.drop([0,1,2,3,4,5],inplace = True)
df
```

```python
#since we have data of countries containing index as mentioned in the country
#we will take the countries below them till next index starts and will classify them

df_temp = pd.DataFrame()
df_fin = pd.DataFrame()

df_temp = df.loc[7:72]
df_temp['HDI'] = 'VERY HIGH HUMAN DEVELOPMENT'
df_fin = pd.concat([df_fin,df_temp])

df_temp = df_temp[0:0]
df_temp = df.loc[74:122]
df_temp['HDI'] = 'HIGH HUMAN DEVELOPMENT'
df_fin = pd.concat([df_fin,df_temp])

df_temp = df_temp[0:0]
df_temp = df.loc[124:167]
df_temp['HDI'] = 'MEDIUM HUMAN DEVELOPMENT'
df_fin = pd.concat([df_fin,df_temp])

df_temp = df_temp[0:0]
df_temp = df.loc[169:]
df_temp['HDI'] = 'LOW HUMAN DEVELOPMENT'
df_fin = pd.concat([df_fin,df_temp])

df_fin
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')

imp_mean.fit(df_fin.iloc[:,2:6])
df_fin.iloc[:,2:6] = imp_mean.transform(df_fin.iloc[:,2:6])
df_fin
```

```python
X = df_fin.iloc[:,[2,3,4,5]]
Y = df_fin.iloc[:,-1]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
nor = preprocessing.Normalizer()
#le.fit(Y)
le.fit_transform(Y)
nor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.4, random_state = 1)
```

```python
#buliding SVC model
df_score = pd.DataFrame()
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,confusion_matrix,ConfusionMatrixDisplay
ker = ['linear','poly','rbf','sigmoid']


for k in ker:
    
    scores = []
    model_SVC = SVC(kernel = k)
    model_SVC.fit(x_train,y_train)
    y_pred = model_SVC.predict(x_test)
    
    
    scores.append(k)
    scores.append(accuracy_score(y_test,y_pred))
    scores.append(precision_score(y_test,y_pred,average = 'weighted',zero_division=0))
    scores.append(recall_score(y_test,y_pred,average = 'weighted',zero_division=0))
    scores.append(f1_score(y_test,y_pred,average = 'weighted',zero_division=0))
    df_score = pd.concat([df_score,pd.Series(scores)], axis =1)
    
    cm = confusion_matrix(y_test,y_pred)
    cmdisp = ConfusionMatrixDisplay(confusion_matrix = cm)
    cmdisp.plot()
    plt.title(k)
    plt.show()
    
    
df_score = df_score.T
df_score.columns = ['Class','Accuracy','Precision','Recall','F1Score']
```
